# 박스플롯(boxplot), 바이올린플롯(violinplot)
- 사분위수(25%, 50%, 75%) 와 함께 최솟값, 최댓값도 함께 표현
- 이상치(outlier)도 함께 표현

```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

# 1. 박스플롯
- 정확한 사분위수 정보가 필요할 때
- 이상치 식별이 중요할 때
- 여러 그룹같 중앙값을 비교 할때

```python
sns.boxplot(x=df['size'], y=df['tip'])
```

# 2. 바이올릿플롯
- 데이터의 분포 모양이 중요할 때
- 다봉분노나 치우침 확인
- 밀도 정보가 필요한 분석

```python
sns.violinplot(x=df['size'], y=df['tip'])
```
