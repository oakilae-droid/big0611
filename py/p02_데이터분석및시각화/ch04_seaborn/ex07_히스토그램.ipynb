# 히스토그램(histplot)
- 막대그래프와 차이점: 막대들 사이에 간격이 없다.

tips 데이터 활용

```python
import seaborn as sns
df = sns.load_dataset("tips")
```

```python
# 1. 기본 히스토그램 생성
sns.histplot(data=df['total_bill'])
```

## 기본 문법
### sns.histplot(data)
- data: 데이터프레임 또는 다차원 배열
- hue: 그룹필할 변수(색상) -> 범례
- bins: 막대의 개수 -> 구간
- kde: Kernel Density Estimation(커널 밀도 추정)

```python
# 2. bins(구간) 조정
sns.histplot(data=df['total_bill'], bins=5)
```

```python
# 3. KDE(커널 밀도 추정)
sns.histplot(data=df['total_bill'], kde=True)
```

# 분포 시각화를 위한 다른 방법

```python
# 1. kde plot
sns.kdeplot(data=df['total_bill'])
```

```python
# 2. rug plot
# - 바닥에 까는 카페트(rug) 모양처럼 작은 실 모양의 
sns.rugplot(data=df['total_bill'])
```

```python
# 3. 히스토그램 vs 카운트 플롯
# - 히스토그램: 연속형 변수의 분포 확인
sns.histplot(data=df['day'])
```

```python
# 카운트플롯
sns.countplot(x=df['day'])
```
