# 히트맵(heatmap)
- 여러가지 변수를 한 번에 비교할 때
- 2차원 격자 모양으로 나뉜 각각의 칸에 데이터의 값을 색상으로 표시
- 3차원 데이터(x축, y축, 값)을 2차원 평면에 효과적으로 표현
- x축, y축 값: 범주형 데이터(실수형 데이터는 사용x)
- 격자 안의 값: 숫자 형태의 변수만 가능
- 표시할 값: 평균값, 최댓값, 최솟값, 합계, 개수 등

### tips 데이터 활용

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
# 1. 기본 히트맵 생성
pivot_df = df.pivot_table('tip', 'day', 'size', observed=True)
pivot_df
```

```python
# 1. 기본 히트맵 생성
sns.heatmap(pivot_df)
```

```python
# 2. 히트맵 값 표시
sns.heatmap(pivot_df, annot=True)
```

```python
# 카테고리 순서 지정
day_order = ['Thur', 'Fri', 'Sat', 'Sun']
size_order = [1, 2, 3, 4, 5, 6]

# 순서가 지정된 피봇 테이블
pivot_ordered = df.pivot_table('tip', 'day', 'size')
pivot_ordered
```

```python
pivot_ordered = pivot_ordered.reindex(day_order)
pivot_ordered
```

```python
pivot_ordered = pivot_ordered.reindex(columns=size_order)
pivot_ordered
```
