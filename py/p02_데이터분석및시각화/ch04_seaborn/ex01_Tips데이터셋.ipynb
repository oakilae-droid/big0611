```python
# Tips 데이터셋
## seaborn 내장 데이터넷
import seaborn as sns
import pandas as pd
```

```python
# 1. 사용 가능한 데이터셋 확인
sns.get_dataset_names()
```

```python
# 2. 데이터 로드 방법
# sns.load_dataset('데이터셋이름')
df = sns.load_dataset('tips')
df.head()
```

```python
# 3. 데이터 구조
df.info()
```

```python
# 4. 데이터 분포
df.describe()
```

```python
# 5. 범주형 데이터의 탐색
# 성별
# df['컬럼명'].unique(): 유일한 값들
df['sex'].unique()
```

```python
# df.['컬럼명'].value_counts(): 각 값의 개수
df['sex'].value_counts()
```

```python
# 전체 범주형 데이터 요약
# 모든 범주형 변수의 고유값 개수
categorical_columns = df.select_dtypes(include=['category']).columns
categorical_columns
```

```python
# df['컬럼명'].nunique(): 고유값 개수
# df['컬럼명'].unique(): 고유값 리스트
for col in categorical_columns:
    print(f'{col}: {df[col].nunique()}개의 고유값')
    print(f'   값: {list(df[col].unique())}')
    print()
```
