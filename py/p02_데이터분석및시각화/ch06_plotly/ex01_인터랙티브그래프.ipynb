## 첫 번째 그래프 미리보기

```python
# 간단한 데이터 생성
data = {
    'x': [1, 2, 3, 4],
    'y': [10, 11, 12, 13]
}

# 선 그래프 생성
fig = px.line(data, x = 'x', y = 'y', title='첫 번째 Plotly 그래프')
fig.show()
```
