## 선 그래프(line plot)

```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
# 시계열 데이터 생성
# 2024-01-01 부터 D(Daily) 하루씩 30일간
dates = pd.date_range('2024-01-01', periods=30, freq='D')
# np.random.randn(30): 평균 0, 표준편차 1에 따른 난수 30개
values = np.cumsum(np.random.randn(30)) + 100

time_series = pd.DataFrame({
    '날짜': dates,
    '주가': values
})
time_series
```

```python
fig = px.line(
    time_series,
    x='날짜',
    y='주가',
    title='일일 주가 변동',
    markers=True
)
fig.show()
```
