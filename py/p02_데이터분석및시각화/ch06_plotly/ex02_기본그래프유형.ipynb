# 기본 그래프 유형
## 선 그래프(line plot)

```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
# 시계열 데이터 생성
# 2024-01-01 부터 D(Daily) 하루씩 30일간
dates = pd.date_range('2024-01-01', periods=30, freq='D')
# np.random.randn(30): 평균 0, 표준편차 1에 따른 난수 30개
values = np.cumsum(np.random.randn(30)) + 100

time_series = pd.DataFrame({
    '날짜': dates,
    '주가': values
})
time_series
```

```python
fig = px.line(
    time_series,
    x='날짜',
    y='주가',
    title='일일 주가 변동',
    markers=True
)
fig.show()
```

```python
# 2. 다중 선 그래프
companies = ['삼성전자', 'LG전자', 'SK하이닉스']
stock_data = []

for company in companies:
    for i, date in enumerate(dates):
        price = 100 + np.cumsum(np.random.randn(1))[0] * (i + 1) / 10
        stock_data.append({
            '날짜': date,
            '회사': company,
            '주가': price
        })
df_stocks = pd.DataFrame(stock_data)
df_stocks
```

```python
# 데이터 시각화
fig = px.line(
    df_stocks,
    x='날짜',   # x축 컬럼명
    y='주가',   # y축 컬럼명
    color='회사',   # 그룹별 색상
    title='회사별 주가 비교',   # 그래프 제목
    line_dash='회사'    # 선 스타일
)
fig
```

# 산점도(scatter plot)

```python
# 1. 기본 산점도
# 키와 몸무게 데이터
height_weight = {
    '키': [160, 165, 170, 175, 180, 162, 168, 172, 178, 185],
    '몸무게': [55, 60, 65, 70, 75, 58, 63, 68, 73, 80],
    '성별': ['여성', '여성', '여성', '남성', '남성', '여성', '여성', '남성', '남성', '남성']
}

df_body = pd.DataFrame(height_weight)
df_body
```

```python
fig = px.scatter(df_body, x='키', y='몸무게', title='키와 몸무게의 관계')
fig.show()
```

```python
# 2. 그룹별 산점도
# 성별로 구분한 산점도
fig = px.scatter(
    df_body,    # 데이터 프레임
    x='키', # x축 컬럼 -> x축 값 및 이름
    y='몸무게', # y축 컬럼 -> y축 값 및 이름
    color='성별',   # 그룹별 색상 -> 범례
    title='성별에 따른 키와 몸무게 관계',   # 그래프 제목
    size_max=15 # 마커의 최대 크기
)

# 축 제목 변경
fig.update_layout(
    xaxis_title='키 (cm)',
    yaxis_title='몸무게 (kg)'
)

fig.show()
```

```python
# 3. 버블 차트
# 크기까지 고려한 버블 차트
# 데이터 생성(직접, 크롤링: 웹 데이터 가져오기)
countries = {
    '국가': ['한국', '일본', '중국', '미국', '독일'],
    'GDP': [1930, 4380, 20850, 32380, 5450],
    '인구': [51, 122, 1420, 342, 83],
    '대륙': ['아시아', '아시아', '아시아', '북미', '유럽']
}

# 데이터 분석(전처리 -> 테이터 다듬기, 통계)
df_countries = pd.DataFrame(countries)
df_countries
```

```python
# 머신러닝(기계 학습)
# 데이터 시각화
fig = px.scatter(
    df_countries,
    x='GDP',
    y='인구',
    size='GDP',
    color='대륙',
    title='국가별 GDP와 인구 관계',
    size_max=60,
    labels={    # 축제목
        'GDP': 'GDP (억 달러)',
        '인구': '인구 (백만명)'
        }
)

# fig.update_layout(
#     xaxis_title = 'GDP (억 달러)',
#     yaxis_title = '인구 (백만명)'
# )

fig.show()
```

# 막대 그래프(bar chart)

```python
# 1. 세로 막대 그래프
# 데이터 생성
# 과목별 점수 데이터
subjects = {
    '과목': ['수학', '영어', '국어', '과학', '사회'],
    '점수': [85, 92, 78, 88, 82]
}

subjects
```

```python
# 데이터 분석
df_scores = pd.DataFrame(subjects)
df_scores
```

```python
# 데이터 시각화
fig = px.bar(
    df_scores,
    x='과목',
    y='점수',
    color='점수',
    color_continuous_scale='Blues'
)
fig.show()
```

```python
# 2. 가로 막대 그래프
# 가로 막대 그래프로 표시
fig = px.bar(df_scores, x='점수', y='과목',
             title='과목별 점수 (가로형)',
             orientation='h',
             color='점수',
             color_continuous_scale='Viridis')
fig.show()
```

```python
# 3. 그룹 막대 그래프
# 학급별 과목 점수 비교
# 데이터 생성
class_scores = {
    '과목': ['수학', '영어', '국어', '과학'] * 3,
    '점수': [85, 92, 78, 88, 82, 88, 75, 90, 80, 85, 82, 86],
    '학급': ['1반'] * 4 + ['2반'] * 4 + ['3반'] * 4
}
```

```python
# 데이터 분석
df_class = pd.DataFrame(class_scores)
df_class
```

```python
# 데이터 시각화
fig = px.bar(
    df_class,
    x='과목',
    y='점수',
    color='학급',
    title='학급별 과목 점수 비교',
    barmode='group' # 'stack'으로 변경하면 누적 막대
)

fig.update_layout(
    xaxis_title = '과목',
    yaxis_title = '점수'
)

fig.show()
```

# 히스토그램(histogram)

```python
# 1. 기본 히스토그램
# 시험 점수 분포 데이터
# 데이터 생성
# NumPy의 난수만 고정
np.random.seed(42)
# np.random.normal(평균, 표준편차, 개수)
# 평균 75, 표준편차 15인 정규분포(normal distribution)에서 랜덤 숫자 1000개를 생성
scores = np.random.normal(75, 15, 1000)
# 0~100 사이로 제한
# np.clip(배열, 최소값, 최댓값)
scores = np.clip(scores, 0, 100)
scores
```

```python
# 데이터 분석
df_hist = pd.DataFrame({'점수': scores})
df_hist
```

```python
# 데이터 시각화
fig = px.histogram(
    df_hist,
    x='점수',
    title='시험 점수 분포',
    nbins=20,
    marginal='rug'
)

fig.update_layout(
    xaxis_title='점수',
    yaxis_title='빈도'
)

fig.show()
```

```python
# 2. 그룹별 히스토그램
# 성별 키 분포 비교
height_data = []
np.random.seed(42)
```

```python
# 남성 키 데이터 (평균 175cm)
male_heights = np.random.normal(175, 6, 500)
for height in male_heights:
    height_data.append({'키': height, '성별': '남성'})

# 여성 키 데이터 (평균 162cm)
female_heights = np.random.normal(162, 5, 500)
for height in female_heights:
    height_data.append({'키': height, '성별': '여성'})

df_heights = pd.DataFrame(height_data)

fig = px.histogram(df_heights, x='키', 
                   color='성별',
                   title='성별 키 분포 비교',
                   marginal='box',  # 박스플롯 추가
                   opacity=0.7,
                   barmode='overlay')

fig.update_layout(
    xaxis_title="키 (cm)",
    yaxis_title="빈도"
)

fig.show()
```

```python
# 캔들스틱 차트 생성
import plotly.graph_objects as go
from datetime import datetime, timedelta

# 주가 데이터 생성
dates = [datetime.now() - timedelta(days=x) for x in range(30, 0, -1)]
np.random.seed(42)

ohlc_data = []
price = 100

for date in dates:
    open_price = price
    high_price = open_price + np.random.uniform(0, 5)
    low_price = open_price - np.random.uniform(0, 5)
    close_price = open_price + np.random.uniform(-3, 3)

    ohlc_data.append({
        '날짜': date,
        '시가': open_price,
        '고가': high_price,
        '저가': low_price,
        '종가': close_price
    })

    price = close_price

df_ohlc = pd.DataFrame(ohlc_data)
df_ohlc
```

```python
# 캔들스틱 차트 생성
fig = go.Figure(data=go.Candlestick(
    x=df_ohlc['날짜'],
    open=df_ohlc['시가'],
    high=df_ohlc['고가'],
    low=df_ohlc['저가'],
    close=df_ohlc['종가'],
    name='주가'
))

fig.update_layout(
    title='일별 주가 캔들스틱 차트',
    xaxis_title='날짜',
    yaxis_title='가격',
    xaxis_rangeslider_visible=False  # 하단 범위 슬라이더 제거
)

fig.show()
```

```python
# 거래량과 함께 표시
# 거래량 데이터 추가
df_ohlc['거래량'] = np.random.randint(1000, 10000, len(df_ohlc))

# 서브플롯 생성
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.03,
    subplot_titles=('주가', '거래량'),
    row_width=[0.2, 0.7]
)

# 캔들스틱 차트
fig.add_trace(
    go.Candlestick(
        x=df_ohlc['날짜'],
        open=df_ohlc['시가'],
        high=df_ohlc['고가'],
        low=df_ohlc['저가'],
        close=df_ohlc['종가'],
        name='주가'
    ),
    row=1, col=1
)

# 거래량 막대 차트
fig.add_trace(
    go.Bar(
        x=df_ohlc['날짜'],
        y=df_ohlc['거래량'],
        name='거래량',
        marker_color='rgba(158,202,225,0.8)'
    ),
    row=2, col=1
)

fig.update_layout(
    title='주가와 거래량',
    xaxis_rangeslider_visible=False
)

fig.show()
```
