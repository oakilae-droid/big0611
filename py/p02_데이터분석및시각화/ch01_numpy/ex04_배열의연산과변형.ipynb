```python
# 배열의 연산과 변형
import numpy as np

# 배열 생성
arr = np.arange(1, 5)
print(arr)
```

```python
## 벡터화 연산(Vectorized Operations)
print(arr + 2)
```

```python
print(arr * 2)
```
