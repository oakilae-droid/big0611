```python
# 참고
# *는 모든것
from numpy import *
```

```python
# 리스트의 항목에 2 를 곱하기
# 리스트 연산 -> 반복문 필요!
list_data = [1, 2, 3, 4]
result_list = []
for item in list_data:
    result_list.append(item * 2)
print(result_list)
```

```python
# 배열 연산 - > 반복문 불필요!
# 벡터화 연산
array_data = np.array([1, 2, 3, 4])
result_array = array_data * 2
print(result_array)
```
