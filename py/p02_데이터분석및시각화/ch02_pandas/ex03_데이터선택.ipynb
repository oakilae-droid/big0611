```python
import pandas as pd
```

## 기본 인덱싱과 슬라이싱

```python
# 예시 데이터 생성
sales_data = pd.DataFrame({
    'product': ['노트북', '마우스', '키보드', '모니터', '스피커'],
    'price': [1200000, 25000, 80000, 350000, 120000],
    'quantity': [10, 50, 30, 15, 25],
    'category': ['전자제품', '주변기기', '주변기기', '전자제품', '주변기기']
})
```

```python
# 단일 컬럼 선택
product_series = sales_data['product']
print(product_series)
```

```python
# 다중 컬럼 선택
price_quantity = sales_data[['product', 'price', 'quantity']]
print(price_quantity)
```

```python
# 3. 행 슬라이싱(첫 3개 행 선택)
first_three = sales_data[0:3]
print(first_three)
```

## 불리언 인덱싱(Boolean Indexing) - 조건부 선택

```python
# 1. 불리언 인덱싱
# df[조건식]
# 가격(price)이 100,000원 이상인 제품 선택
# 조건식: sales_data['price'] >= 100000
expensive_products = sales_data[sales_data['price'] >= 100000]
print(expensive_products)
```

```python
# 2. df.query()
# high_quantity = sales_data[sales_data['quantity'] > 20]
high_quantity = sales_data.query('quantity > 20')
high_quantity
```

## loc과 iloc을 활용한 라벨/위치 기반 선택

```python
# 인덱스 설정
sales_data_indexed = sales_data.set_index('product')
print(sales_data_indexed)
```

```python
# 1. loc: 라벨 기반 선택
notebook_info = sales_data_indexed.loc['노트북', 'price']
print(f"노트북 가격: {notebook_info}")
```

```python
# 2. 여러 제품의 가격과 수량 선택
# df.loc[[행라벨, ...], [열라벨, ...]]: 라벨기반 선택
selected_data = sales_data_indexed.loc[['노트북', '모니터'], ['price', 'quantity']]
print(selected_data)
```

```python
# 3. 첫 번째와 세 번째 행, 두 번째와 세 번째 컬럼
# df.iloc[행index, 열index]
positional_data1 = sales_data.iloc[0:3, 1:3]
positional_data2 = sales_data.iloc[[0, 2], [1, 2]]
positional_data3 = sales_data.iloc[[0, 1, 2], [1, 2]]
print(sales_data)
print('data1: \n', positional_data1)
print('data2: \n', positional_data2)
print('data3: \n', positional_data3)
```

## 다중 조건 선택

```python
# 1. 가격(price)이 5,000원 이상이면서(and, &) 수량(quantity)이 20개 이상인 제품
# 가격 >= 5000 & 수량 >= 20
# sales_data['price'] >= 5000 & sales_data['quantity'] >= 20
filtered_products = sales_data[(sales_data['price'] >= 5000) & (sales_data['quantity'] >= 20)]
print(sales_data)
print(f'\n{filtered_products}')
```

```python
# 2. 전자제품(category) 이거나 가격(price)이 30,000원 이하인 제품
# sales_data['category'] == '전자제품' | sales_data['price'] <= 30000
electronic_or_cheap = sales_data[(sales_data['category'] == '전자제품') | (sales_data['price'] <= 30000)]
print(sales_data)
print(f'\n{electronic_or_cheap}')
```

## 그룹별 데이터 선택

```python
# 카테고리별 그룹화 후 전자제품 그룹만 선택
# df.groupby('컬럼명').get_group('데이터')
electronics_group = sales_data.groupby('category').get_group('전자제품')
print(electronics_group)
```

```python
# 각 카테고리별 평균 가격을 원본 데이터에 추가
sales_data['category_avg_price'] = sales_data.groupby('category')['price'].transform('mean')
print(sales_data)
print(sales_data[['product', 'category', 'price', 'category_avg_price']])
```

## 시계열 데이터 선택

```python
# 시계열 데이터 생성
# pd.date_range('시작날짜', periods=기간, freq='D')
# freq=: 빈도수의 기준, D(Day, 일), W(Week, 주), ME(Month End, 월말), MS(Month Start, 월초)
dates = pd.date_range('2025-12-31', periods=5, freq='D')
time_series_data = pd.DataFrame({
    'date': dates,
    'sales': [100, 150, 200, 120, 180],
    'returns': [5, 8, 12, 6, 9]
})
time_series_data.set_index('date', inplace=True)
# inplace=True: 인덴스 자리에 date로 대체 한다

print(time_series_data)
```

```python
# 1. 특정 날짜 범위 선택
jan_1_to_3 = time_series_data.loc['2026-01-01':'2026-01-03']
print(jan_1_to_3)
```

```python
# 2. 연도 기준 필터링
# 2026년 데이터만 선택
# df.index.year
# df.index.month
# df.index.day
year_2026 = time_series_data[time_series_data.index.year == 2026]
print(year_2026)
```
